```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import mnist

images, labels = mnist.load_mnist_gz()

print(images.shape)
print(labels.shape)

fig, axs = plt.subplots(8, 15, figsize=(15, 8))
for i, ax in enumerate(np.reshape(axs, -1)):
    ax.axis('off')
    ax.imshow(255.-images[i], cmap='gray')
```
